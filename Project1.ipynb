```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
```

```python
sns.set(rc={'figure.figsize':(12,4)})
```

```python
df_airbnb = pd.read_csv("AB_NYC_2019.csv")
df_airbnb.head()
```

```python
df_airbnb["neighbourhood_group"].unique()
```

```python
df_airbnb["neighbourhood_group"].value_counts()
```

```python
df_airbnb["neighbourhood_group"].value_counts().plot.bar().set_title("Neighbourhood Group Counts")
```

```python
sns.barplot(x="neighbourhood_group", y=df_airbnb["neighbourhood_group"].index, color="pink", data=df_airbnb).set_title("density")
```

```python
sns.barplot(x="neighbourhood", y=df_airbnb["neighbourhood"].index, hue="room_type", data=df_airbnb).set_title("density")
```

```python
df_airbnb.info()
```

```python
sns.barplot(x="neighbourhood_group", y="price", hue="room_type", data=df_airbnb).set_title("neighbourhood-price-room")
```

```python
df_airbnb.groupby("room_type").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(4,4))
```

```python
cols = ["number_of_reviews", "reviews_per_month", "availability_365"]
for col in cols:
  plt.figure(figsize=(5,5))
  sns.scatterplot(x=col, y="price", data=df_airbnb)
```

```python
sns.histplot(df_airbnb["latitude"], kde=True)
```

```python
sns.histplot(df_airbnb["latitude"], kde=True, bins=50)
```

```python
sns.histplot(df_airbnb["latitude"])
```

```python
sns.kdeplot(df_airbnb["latitude"])
```

```python
sns.catplot(x="neighbourhood_group", y="price", hue="room_type", kind="point",data=df_airbnb)
plt.title("neighbourhood-price-room")
```

```python
sns.kdeplot(df_airbnb["latitude"], shade=True)
```

```python
(sns.FacetGrid(df_airbnb, hue="neighbourhood_group", height=5, xlim=(0,1000)).map(sns.kdeplot, "price", shade=True).add_legend())
```

```python
plt.title("Correlation Matrix")
sns.heatmap(df_airbnb.corr(), annot=True)
```
